```python
lr_sklearn=LinearRegression()
Xsk=np.reshape(X, (-1, 1))
Ysk=np.reshape(Y, (-1, 1))
lr_sklearn.fit(Xsk,Ysk)
```

```python
m_sk=lr_sklearn.coef_[0][0]
b_sk=lr_sklearn.intercept_[0]
print(f"{m_sk}, {b_sk}")
```

```python
def cost(m, b, X, Y):
    return (1/(2*len(X)))*np.dot(m*X+b-Y, m*X+b-Y)
```

```python
def DcostDm(m, b, X, Y):
    return 1/len(X)*np.dot(m*X + b - Y, X)
def DcostDb(m, b, X, Y):
    return 1/len(X)*np.dot(m*X + b - Y, (X*0)+1) #dirty unreadable code, cus why not, lol, proof of human thought process and lazyness
```

```python
def gradDesc(m, b, X, Y, alpha, itern, printCost=False):
    for _ in range(itern):
        temp=m-alpha*DcostDm(m, b, X, Y)
        b-=alpha*DcostDb(m, b, X, Y)
        m=temp
        if printCost:
            print(cost(m, b, X, Y))
    return m, b
```

```python
mG, bG=gradDesc(0, 0, X, Y, 0.001, 1000)
mG1, bG1=gradDesc(0, 0, X, Y, 0.01, 100)
mG2, bG2=gradDesc(0, 0, X, Y, 1, 100)
```
